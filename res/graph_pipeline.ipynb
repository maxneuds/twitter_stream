# Pipeline Graph

```python
from graphviz import Digraph
g = Digraph('usecases', filename='graph_pipeline.gv', format='png')
g.attr(rankdir='TB', compound='true', newrank='true', dpi='300')
g.node_attr['shape'] = 'box'
g.node_attr['width'] = '2.3'
g.node_attr['height'] = '.65'
g.node_attr['fontname'] = 'hack'

# usecases
g.node('p1', 'Twitter')
g.node('p2', 'Tweepy')
g.node('p3', 'Streamz')
g.node('p4', 'Sink: dict')
g.node('p5', 'Pandas: DataFrame')


# edges
g.edge('p1', 'p2', label='  source')
g.edge('p2', 'p3', label='  filter source')
g.edge('p3', 'p4', label=f'{" " * 10}map: hashtags\n  aggregate: hashtags\n{" " * 3}msink: store')
g.edge('p4', 'p5', label='  sink:save')

# save to file
g.render()

# show in notebook
g.attr(dpi='67')
g
```
